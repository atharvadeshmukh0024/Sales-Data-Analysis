```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the sales dataset
# Replace 'sales_data.csv' with the path to your dataset
data = pd.read_csv('sales_data.csv')
```

```python
# Preview the dataset
print(data.head())
```

```python
# Data Cleaning
# Fill missing values
data.fillna(0, inplace=True)
```

```python
# Convert 'Order Date' to datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# Add a 'Profit Margin' column
data['Profit Margin'] = (data['Profit'] / data['Sales Amount']) * 100
```

```python
# Sales trend over time
monthly_sales = data.groupby(data['Order Date'].dt.to_period('M'))['Sales Amount'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales Trend', figsize=(10, 6))
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# Top 10 products by sales
top_products = data.groupby('Product Name')['Sales Amount'].sum().nlargest(10)
top_products.plot(kind='bar', title='Top 10 Products by Sales', figsize=(10, 6), color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by region
region_sales = data.groupby('Region')['Sales Amount'].sum()
region_sales.plot(kind='pie', autopct='%1.1f%%', title='Sales by Region', figsize=(8, 8))
plt.ylabel('')
plt.show()
```

```python
# Customer segmentation analysis
customer_segments = data.groupby('Customer Segment')['Sales Amount'].sum()
sns.barplot(x=customer_segments.index, y=customer_segments.values, palette='viridis')
plt.title('Sales by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.show()
```
